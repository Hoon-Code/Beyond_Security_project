```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import img_to_array, load_img

data_dir = 'C:/Users/user/Downloads/output/test/'  # 이 디렉토리를 데이터셋이 있는 실제 디렉토리로 변경하세요

# 데이터셋 경로와 이미지를 저장할 리스트 초기화
all_data = []
all_labels = []

# 각 클래스별로 데이터셋의 경로와 라벨을 저장
for filename in os.listdir(data_dir):
    if filename.endswith('.jpg'):  # 이미지 파일인 경우만 처리
        filepath = os.path.join(data_dir, filename)
        # 파일명에서 클래스를 추출하여 라벨로 사용
        class_name = filename.split('_')[0].lower()  # 파일명에서 '_'을 기준으로 앞 부분을 추출하고 소문자로 변환
        if class_name in ['normal', 'violence', 'damage', 'thief']:
            label = ['normal', 'violence', 'damage', 'thief'].index(class_name)
            
            # 이미지를 NumPy 배열로 변환하여 저장
            img = load_img(filepath, target_size=(156, 156))  # 이미지를 224x224 크기로 로드
            img_array = img_to_array(img)
            all_data.append(img_array)
            all_labels.append(label)

# 리스트를 NumPy 배열로 변환
all_data_np = np.array(all_data)/255.0
all_labels_np = np.array(all_labels)

# 데이터셋을 훈련(train) 및 검증(val) 세트로 분할
train_data, val_data, train_labels, val_labels = train_test_split(
    all_data_np, all_labels_np, test_size=0.2, random_state=42, stratify=all_labels_np
)

# 데이터셋 분할 결과 확인
print(f"Total data: {len(all_data_np)}")
print(f"Train data: {len(train_data)}, Train labels: {len(train_labels)}")
print(f"Validation data: {len(val_data)}, Validation labels: {len(val_labels)}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import ModelCheckpoint,EarlyStopping
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
```

```python
import os
import re
from shutil import move

def classify_and_move(file_path, destination_dir):
    # 파일명에서 규칙에 따라 클래스 추출
    class_name = None
    if re.search(r'abnormal', file_path):
        class_name = 'strange'
    elif re.search(r'normal', file_path):
        class_name = 'normal'

    if class_name:
        # 대상 디렉토리에 클래스별 폴더가 없으면 생성
        class_dir = os.path.join(destination_dir, class_name)
        os.makedirs(class_dir, exist_ok=True)

        # 파일을 대상 디렉토리의 클래스별 폴더로 이동
        move(file_path, os.path.join(class_dir, os.path.basename(file_path)))
        print(f"Moved '{file_path}' to '{class_name}' folder.")
    else:
        print(f"Couldn't classify '{file_path}'.")

# 예시: 분류 대상 디렉토리
source_directory = 'C:/Users/user/Downloads/DATA-output/DATA/test'

# 예시: 분류 결과 저장 디렉토리
destination_directory = 'C:/Users/user/Downloads/DATA-output/DATA/test2'

# 분류 대상 디렉토리의 모든 파일에 대해 수행
for filename in os.listdir(source_directory):
    file_path = os.path.join(source_directory, filename)
    if os.path.isfile(file_path):
        classify_and_move(file_path, destination_directory)
```

```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import img_to_array, load_img

data_dir = 'C:/Users/user/Downloads/DATA-output/DATA/test2'  # 이 디렉토리를 데이터셋이 있는 실제 디렉토리로 변경하세요

# 데이터셋 경로와 이미지를 저장할 리스트 초기화
all_data = []
all_labels = []

for class_name in os.listdir(data_dir):
    class_folder = os.path.join(data_dir, class_name)
    if os.path.isdir(class_folder):
        label = 0 if class_name == 'normal' else 1

        for filename in os.listdir(class_folder):
            if filename.endswith('.jpg'):
                filepath = os.path.join(class_folder, filename)

                # 이미지를 NumPy 배열로 변환하여 저장
                img = load_img(filepath, target_size=(156, 156))  # 이미지를 156x156 크기로 로드
                img_array = img_to_array(img)
                all_data.append(img_array)
                all_labels.append(label)

# 리스트를 NumPy 배열로 변환
all_data_np = np.array(all_data) / 255.0
all_labels_np = np.array(all_labels)

# 데이터셋을 훈련(train) 및 검증(val) 세트로 분할
train_data, val_data, train_labels, val_labels = train_test_split(
    all_data_np, all_labels_np, test_size=0.2, random_state=42, stratify=all_labels_np
)

# 데이터셋 분할 결과 확인
print(f"Total data: {len(all_data_np)}")
print(f"Train data: {len(train_data)}, Train labels: {len(train_labels)}")
print(f"Validation data: {len(val_data)}, Validation labels: {len(val_labels)}")
```
